```r
require(Hmisc)
require(repr)
require(stats)
require(ggplot2)
require(ggpubr)
```

## Exercise 1.1

The corresponding R code is as follows:

```r
exhibit.1.2 <- read.table("../datasets/larain.dat", header = TRUE)[["larain"]]

scatter <- ggplot()
scatter <- scatter + geom_point(aes(x = Lag(exhibit.1.2), y = exhibit.1.2), 
                                color = "blue", na.rm=T)
scatter <- scatter + labs(x = "Previous Year Inches", y = ("Inches")) 
scatter <- scatter + theme_bw() + theme(aspect.ratio = 1)
scatter
```

## Exercise 1.2

The corresponding R code is as follows:

```r
exhibit.1.3 <- read.table("../datasets/color.dat", header = TRUE)[["color"]]
scatter <- ggplot()
scatter <- scatter + geom_line(aes(x = seq_along(exhibit.1.3), y = exhibit.1.3), 
                               color = "blue")
scatter <- scatter + geom_point(aes(x = seq_along(exhibit.1.3), y = exhibit.1.3), 
                               color = "blue")
scatter <- scatter + labs(x = "Batch", y = ("Color Property")) 
scatter <- scatter + theme_bw() + theme(aspect.ratio = 1)
scatter
```
