```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Question 2
def gaussian_distribution(x, mu, sigma):
    g_x = 1 / (sigma * np.sqrt(2 * np.pi)) *  \
        np.exp((-1) * np.power(x - mu, 2) / (2 * np.power(sigma, 2)))
    return g_x

x = np.linspace(130,210,1000)
plt.scatter(x, gaussian_distribution(x, 155, 3), s= 1.0)
plt.grid(True)
plt.show()
```

```python
# Question 3
x = np.linspace(0,10,1000)
plt.scatter(x, gaussian_distribution(x, 3, 3), s= 1.0)
plt.grid(True)
plt.show()
```
